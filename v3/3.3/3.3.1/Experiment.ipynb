```python
import random
import matplotlib.pyplot as plt
import numpy as np

def plot_freq_matrix(m):
    print(m)
    data = np.array(m)
    plt.imshow(data, cmap='viridis')  # 'viridis' is a common colormap
    plt.colorbar()  # Add a colorbar to show the mapping of values to colors
    plt.show()
    
r = []
for i in range(1,20001):
    r += [random.randint(0,9)]

m = [[0 for _ in range(10)] for _ in range(10)]

for i in range(0, 10000):
    x= i*2
    y = i*2+1
    m[r[x]][r[y]]+=1
plot_freq_matrix(m)


m = [[0 for _ in range(10)] for _ in range(10)]

for i in range(0, 19999):
    x= i
    y = i+1
    m[r[x]][r[y]]+=1
plot_freq_matrix(m)
```
